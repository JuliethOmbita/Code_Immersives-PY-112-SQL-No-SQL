<h2>Exercise</h2>
Find all numbers from 1000 to 3000 that are even numbers and when you add the digits <br>
The sum is divisible by 5<br>
Also output the total number of items found<br>
for example 1004 is even and the sum of the digits is 5 which is divisible by 5

```python
# step 1 - iterate overthe domain
for i in range(1000,3001):
    print(i)
```

```python
# step 2 - get the even numbers
for i in range(1000,3001):
    if i % 2 == 0:
        print(i)
```

```python
# step 3 - Turn the number into a string
for i in range(1000,3001):
    if i % 2 == 0:
        print(i, str(i))
```

```python
# step 4 - Sum the digits
for i in range(1000,3001):
    if i % 2 == 0:
        sum_str = [int(x) for x in str(i)]  # A list of integers
        print(i, str(i), sum_str)
```

```python
# step 5 - is the sum of the list of integers divisible by 5
for i in range(1000,3001):
    if i % 2 == 0:
        sum_str = sum([int(x) for x in str(i)])  # sum
        if sum_str % 5 == 0:
            print(i, str(i), sum_str)
```

Take the following words and sum up their ascii values<br>
For the first word 'Love'<br>
L = 76<br>
o = 111<br>
v = 118<br>
e = 101<br>
The sum is 406
words = ['Love','Python','Dictionary','List','Tuple','Set','Ternary']<br>
Return the word and the sum as a list of tuples:<br>
[('Love', 406), ('Python',...

```python
# step 1 - iterate over the domain
words = ['Love','Python','Dictionary','List','Tuple','Set','Ternary']
for word in words:
    print(word)
```

```python
# step 2 - iterate over the word add the ord next to it
words = ['Love','Python','Dictionary','List','Tuple','Set','Ternary']
for word in words:
    print(word)
    for ch in word:
        print(ch,ord(ch))
```

```python
# step 3 - Create a total
words = ['Love','Python','Dictionary','List','Tuple','Set','Ternary']
for word in words:
    print(word)
    tot = 0
    for ch in word:
        tot += ord(ch)
        print(ch,ord(ch),tot)
```

```python
# step 4 - Add my list of tuples
words = ['Love','Python','Dictionary','List','Tuple','Set','Ternary']
word_out = []
for word in words:
    print(word)
    tot = 0
    for ch in word:
        tot += ord(ch)
        print(ch,ord(ch),tot)
    x = (word,tot)
    word_out.append(x)
print('='*35)
print(word_out)
```

<h2>Exercise</h2>
Calculate the sum of the factors of all composite number from 100 to 200<br>
Output the number, the sum of the factors, and the factor to the console

```python
# step 1 - iterate over the domain
for i in range(100,201):
    print(i)
```

```python
def factors(num):
    if num < 1:
        prime = False
        factors = []
    else:
        factors = [1]
        prime = True
        for i in range(2,num):
            if num % i == 0:
                prime = False
                factors.append(i)
    return factors + [num] if not prime else []

for i in range(25):
    ans = factors(i)
    if len(ans) > 0:
        print(i,sum(ans),ans)
print('='*25)
for i in range(100,300):
    ans = factors(i)
    if len(ans) > 0:
        print(i,sum(ans),ans)
```
