<h2>Exercise - Count the '9'  </h2>
<br>
Create a program that counts the number of times a nine ('9') appears in a number from 100 to 1,000. 
<br>For example the number 199 has
2 nines in the number.

```python
def count_9(num1, num2):
    for num in range (num1, num2):
        num = str(num)
        number_9 = 0
        if '9' in num:
            number_9 = num.count('9')
            print (f'The number {num} has {number_9}')

count_9(100, 1000)
```
