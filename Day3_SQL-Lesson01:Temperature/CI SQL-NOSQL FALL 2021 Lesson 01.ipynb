<h2>A database holds zero or more tables</h2>

````python
# Create a cursor to access the database
c = conn.cursor()
# This is equivalent to creating a filehandle when using the 'open' function
dir(c)
````

````python
c.execute('''CREATE TABLE customers (
    first_name TEXT,
    last_name TEXT,
    address1 TEXT,
    address2 TEXT,
    city TEXT,
    state TEXT,
    zip TEXT
)''')
conn.commit() # <=== commit our table changes to the database !!!
conn.close()  # Close the connection
'''
WARNING !!!! If the table was created already it is going to give you an error
'''
````

````python
conn = sqlite3.connect('customer.db')
c = conn.cursor()
c.execute('''CREATE TABLE IF NOT EXISTS customers (
    first_name TEXT,
    last_name TEXT,
    address1 TEXT,
    address2 TEXT,
    city TEXT,
    state TEXT,
    zip TEXT
)''')
conn.commit() # <=== commit our table changes to the database !!!
conn.close()  # Close the connection
````
