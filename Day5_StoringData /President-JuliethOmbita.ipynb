```python
import datetime
import math
import re

def calc_years_in_office(beg_date,end_date):
    """
    beg_date: Text version of the begin term MM/DD/YYYY
    end_date: Text version of the end term MM/DD/YYYY
    returns a integer representation of the diff between end_date - beg_date
    """
    m1,d1,y1 = beg_date.split('/')
    m2,d2,y2 = end_date.split('/')
    begin_term = datetime.date(int(y1),int(m1),int(d1))
    end_term = datetime.date(int(y2),int(m2),int(d2))
    return math.ceil(((end_term - begin_term).days)/365.25)

with open('us_presidents_unordered.txt','r') as myFile:

    # 1 - Add the data to a data structure
    headLine = myFile.readline()
    content = myFile.readlines()
    # 2 - Separate the dat into the following: presidential_order, first_name, middle_name, last_name, begin_term, end_term
    clean_content = []
    order_id = []
    president_List = []

    for presd in (content):
        # Clean the data
        cln = presd.replace(".","").replace("'","").replace("\n","")
        cln = re.split(' |,', cln)
        #Finding the middle name or adding a string to to cover the spot of the middle name:
        if len(cln) == 4:
            cln.insert(1," ")
        #Create a Id-list:    
        order_id.append(cln[0])
        president_List.append(cln)
        cln.pop(0)
        #Finding the years in office    
        term_start = (cln[-2])
        term_end = (cln[-1])
        yrs_in_office = calc_years_in_office(term_start,term_end)
        cln.append(str(yrs_in_office))
        clean_content.append(cln)
    

    order_id = [int(i) for i in order_id]
    print('++')

    # 3 - Create a report to disk with the following in presidential order: order, "last_name, first_name middle_initial, begin term, end_term, years in office
    report = zip (order_id, clean_content)
    report = dict(report)
    presidents = sorted(report.items(), key = lambda kv: kv[0])
    print('+'*35)
    # print(presidents)
    # print(type(presidents))
```
