```python
import sqlite3
import os

os.chdir(r'/Users/julis_478/Documents/Python Immersives/PY-112 SQL, No SQL/Classes/Day12')

conn = sqlite3.connect('employee_table.db')  # <== create the database
c = conn.cursor()
c.executescript('''
    create table IF NOT EXISTS employees(
        emp_id TEXT,
        ms_id TEXT, 
        last_name TEXT,
        first_name TEXT,
        hire_date TEXT,
        gender TEXT,
        dept_id TEXT,
        manager_id INT,
        status TEXT,
        region TEXT,
        sick_days_alloted INTEGER,
        sick_day1 INTEGER,
        sick_day2 INTEGER,
        sick_day3 INTEGER
        
    );
    DELETE FROM employees;

    create table IF NOT EXISTS marital_status_code(
        id TEXT,
        description TEXT
    );
    DELETE FROM marital_status_code;

    create table IF NOT EXISTS depts(
        id TEXT,
        description TEXT
    );
    DELETE FROM depts;

    create table IF NOT EXISTS sick_days(
        emp_id TEXT,
        days_out TEXT
    );
    DELETE FROM sick_days;
    ''')
print('List of tables: ')
for row in c.execute("""SELECT name FROM sqlite_master"""):
    print(row)

with open('employee_data.csv') as f:
    headLine = f.readline()
    content = f.readlines()
    for emp in (content):
        line = emp.replace("'",'').replace("\n",'').split(',')
        mstatus_id = 'M' if line[1] == 'Married' else 'D' if line[1] == 'Divorced' else 'S' if line[1] == 'Single' else 'W' if line[1] == 'Widowed' else 'none'
        gender_id = 'M' if line[5] == 'Male' else 'F' if line[5] == 'Female' else 'UD'
        line_elements = (line[0], mstatus_id, line[2], line[3], line[4], gender_id, line[6], line[8], line[9], line[11], line[12], line[13], line[14], line[15])
        c.execute('''INSERT INTO employees VALUES (?,?,?,?,?,?,?,?,?,?,?,?,?,?)''',  line_elements)
        conn.commit()
        if not (int (line[12]) == 0):
            c.execute('''INSERT INTO sick_days VALUES (?,?)''', (line[0], line[12]))
            conn.commit()

marrige_code = [('M', 'Married'), ('D', 'Divorced'), ('S', 'Single'), ('W', 'Widowed')]
for i in marrige_code:
    c.execute('''INSERT INTO marital_status_code VALUES (?,?)''', i)
    conn.commit()

dept_code = [('HR', 'Human Resources'), ('IT', 'Technology'), ('FIN', 'Finance'), ('ACC', 'Accounting'), ('WHS', 'WHServices'), ('EX', 'EX')]
for i in dept_code:
    c.execute('''INSERT INTO depts VALUES (?,?)''', i)
    conn.commit()
```
