```python
import os

import sqlite3
filePath = 'C:\\Projects\\SQL_NOSQL\\'
os.chdir(filePath)
conn2 = sqlite3.connect("ABC_corp2.db")
curr = conn2.cursor()
curr.executescript('''
    create table IF NOT EXISTS allData(
        emp_id TEXT,
        marital_status TEXT,
        last_name TEXT,
        first_name TEXT,
        hire_date TEXT,
        gender TEXT,
        dept_id TEXT,
        dept_name TEXT,
        manager_id TEXT,
        status TEXT,
        term_date TEXT,
        region TEXT,
        sick_days_accumulated TEXT,
        sick_day1 TEXT,
        sick_day2 TEXT,
        sick_day3 TEXT
    );
    DELETE FROM allData;
''')
conn.commit()
curr2 = conn2.cursor()
curr2.executescript('''
    create table IF NOT EXISTS marriage_codes(
        code TEXT,
        description TEXT
    );
    DELETE FROM marriage_codes;
    create table IF NOT EXISTS depts (
        code TEXT,
        description TEXT
    );
    DELETE FROM depts;    
    
    create table IF NOT EXISTS sick_days (
        emp_id TEXT,
        sick_day TEXT
    );
    DELETE FROM sick_days;     
    
    create table IF NOT EXISTS regions (
        code TEXT,
        description TEXT
    );
    DELETE FROM regions;    
''')
conn2.commit()
# ============================================
# filePath = r'C:\Projects\Code immersives\SQLite\Normalization_data.txt'
with open(filePath+'Normalization_data.txt','r') as fh:
    line = fh.readline()
    i = 0
    while line:
        if i > 0:
            # print(line)
            empData = line.split(',')
            # print(empData)
            emp_id = empData[0].strip("'")
            marital_status = empData[1].strip("'")
            last_name = empData[2].strip("'")
            first_name = empData[3].strip("'")
            hire_date = empData[4].strip("'")
            gender = empData[5].strip("'")
            dept_id = empData[6].strip("'")
            dept_name = empData[7].strip("'")
            manager_id = empData[8].strip("'")
            status = empData[9].strip("'")
            term_date = empData[10].strip("'")
            region = empData[11].strip("'")
            sick_days_accumulated = empData[12].strip("'")
            sick_day1 = empData[13].strip("'")
            sick_day2 = empData[14].strip("'")
            sick_day3 = empData[15].strip("'")
            # populate the table
            sql = f"""INSERT INTO allData (emp_id,marital_status,last_name,first_name,hire_date,gender,dept_id,dept_name,
            manager_id,status,term_date,region,sick_days_accumulated,sick_day1,sick_day2,sick_day3) 
            VALUES ('{emp_id}','{marital_status}','{last_name}','{first_name}','{hire_date}','{gender}','{dept_id}','{dept_name}',
            '{manager_id}','{status}','{term_date}','{region}','{sick_days_accumulated}','{sick_day1}','{sick_day2}','{sick_day3}');"""
            print(sql)
            conn.commit()
            curr.execute(sql)
        line = fh.readline()
        i += 1
    curr.execute('''SELECT DISTINCT CASE marital_status 
    WHEN 'Married' THEN 'M'
    WHEN 'Divorced' THEN 'D'
    WHEN 'Single' THEN 'S'
    WHEN 'Widowed' THEN 'W'
    else 'X' END code, marital_status FROM allData''')
    rows = curr.fetchall()
    for row in rows:
        print(row[0],row[1]) 
        # Build the SQL and populate the marriage_codes table
        sql = f"""INSERT INTO marriage_codes (code,description) 
        VALUES ('{row[0]}','{row[1]}');"""
        print(sql)
        curr2.execute(sql)
        conn2.commit()
    # Build the dept table
    curr.execute('''SELECT DISTINCT dept_id, dept_name FROM allData''')
    rows = curr.fetchall()
    for row in rows:
        print(row[0],row[1]) 
        # Build the SQL and populate the dept codes table
        sql = f"""INSERT INTO depts (code,description) 
        VALUES ('{row[0]}','{row[1]}');"""
        print(sql)
        curr2.execute(sql)
        conn2.commit()    
    # Build the sick day table
    curr.execute('''SELECT emp_id,sick_day1,sick_day2,sick_day3 FROM allData WHERE LENGTH(TRIM(sick_day1))> 0''')
    rows = curr.fetchall()
    for row in rows:
        # print(row['emp_id'],row['sick_day1'],row['sick_day2'],row['sick_day3']) 
        # Build the SQL and populate the sick_days table
        sql = ''
        if len(row['sick_day1'].strip()) > 0:
            sql = f"""INSERT INTO sick_days (emp_id,sick_day) VALUES ('{row[0]}','{row['sick_day1'].strip()}');"""
            curr2.execute(sql)
            conn2.commit()
            print(sql)
        #
        if len(row['sick_day2'].strip()) > 0:
            sql = f"""INSERT INTO sick_days (emp_id,sick_day) VALUES ('{row[0]}','{row['sick_day2'].strip()}');"""
            curr2.execute(sql)
            conn2.commit()
            print(sql)
        #
        if len(row['sick_day3'].strip()) > 0:
            sql = f"""INSERT INTO sick_days (emp_id,sick_day) VALUES ('{row[0]}','{row['sick_day3'].strip()}');"""        
            curr2.execute(sql)
            conn2.commit()
            print(sql)
        #
    curr.execute('''SELECT DISTINCT CASE region 
    WHEN 'NORTH EAST' THEN 'NE'
    WHEN 'NORTH WEST' THEN 'NW'
    WHEN 'NORTH' THEN 'NO'
    WHEN 'SOUTH EAST' THEN 'SE'
    WHEN 'SOUTH WEST' THEN 'SW'
    WHEN 'SOUTH' THEN 'SO'
    WHEN 'EAST' THEN 'EA'
    WHEN 'WEST' THEN 'WE'
    else 'X' END code, region FROM allData''')
    rows = curr.fetchall()
    for row in rows:
        print(row[0],row[1]) 
        # Build the SQL and populate the marriage_codes table
        sql = f"""INSERT INTO regions (code,description) 
        VALUES ('{row[0]}','{row[1]}');"""
        print(sql)
        curr2.execute(sql)
        conn2.commit()        
curr2.close()
```

<h2>Union and Union All</h2>

```python
import os
import sqlite3

filePath = 'C:\\Projects\\SQL_NOSQL\\'
os.chdir(filePath)
conn2 = sqlite3.connect("ABC_corp2.db")
curr2 = conn2.cursor()
print('='*40) 
curr2.execute('''SELECT first_name, last_name,marital_status FROM allData 
              WHERE marital_status='Divorced'
              UNION
              SELECT first_name, last_name,marital_status FROM allData 
              WHERE marital_status='Widowed' 
              ''')
tables = curr2.fetchall()
for table in tables:
    print(table)
print('='*40)    
curr2.execute('''SELECT first_name, last_name,marital_status FROM allData 
              WHERE marital_status='Divorced'
              UNION ALL
              SELECT first_name, last_name,marital_status FROM allData 
              ''')
tables = curr2.fetchall()
for table in tables:
    print(table)    
print('='*40)
# UNION eliminated duplicates
curr2.execute('''SELECT first_name, last_name,marital_status FROM allData 
              WHERE marital_status='Divorced'
              UNION
              SELECT first_name, last_name,marital_status FROM allData 
              WHERE marital_status='Divorced' 
              ''')
tables = curr2.fetchall()
for table in tables:
    print(table)
print('='*40)   
# UNION ALL includes duplicates
curr2.execute('''SELECT first_name, last_name,marital_status FROM allData 
              WHERE marital_status='Divorced'
              UNION ALL
              SELECT first_name, last_name,marital_status FROM allData 
              WHERE marital_status='Divorced' 
              ''')
tables = curr2.fetchall()
for table in tables:
    print(table)
print('='*40)  
curr2.close()
```
