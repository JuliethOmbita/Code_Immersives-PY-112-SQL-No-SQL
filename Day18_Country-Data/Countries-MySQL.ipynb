```python
import sqlite3
import os

os.chdir(r'/Users/julis_478/Documents/Python Immersives/PY-112 SQL, No SQL/Classes/Day18')

conn = sqlite3.connect('Country_info.db')  # <== create the database
c = conn.cursor()

# 1 - Calculate the population range
c.execute('SELECT population_c FROM population ORDER BY population_c ASC')
rows = c.fetchall()
print(f'1 - Calculate the population range: ({rows[0][0]}, {rows[-1][0]})')

# 2 - Which countries are in the bottom 20% of population
max_pop = rows[-1][0]
twenty_pop = round((max_pop*20)/100)
c.execute(f'SELECT country FROM population WHERE population_c < {twenty_pop}')
rows = c.fetchall()
print('2 - Which countries are in the bottom 20% of population:')
for row in rows:
    print(f' - {row[0]}')

# 3 - What is the total population of the bottom 50% of countries
fifty_pop = round((max_pop*50)/100)
c.execute(f'SELECT population_c FROM population WHERE population_c < {fifty_pop}')
rows = c.fetchall()
fit_pop = [row[0] for row in rows]
print (f'3 - What is the total population of the bottom 50% of countries: {sum(fit_pop)}')

conn.close()
```

```python
import statistics
c.execute(f'SELECT population_c FROM population')
rows = c.fetchall()

# 4 - What is the mean population
c.execute(f'SELECT sum(population_c)/count(*) FROM population')
rows = c.fetchall()
mean_p = rows[0][0]
print (f'4 - What is the mean population {mean_p}')
```

```python
# 5 - Using the statistics library what is the standard deviation of the population data
import statistics
c.execute(f'SELECT population_c FROM population')
rows = c.fetchall()
pop = [row[0] for row in rows]
std_derv = round(statistics.stdev(pop),2)
print (f'5 - What is the standard deviation of the population data: {std_derv}')

# 6 - How many countries are within 1 standard deviation of the mean
c.execute(f'SELECT country FROM population WHERE population_c > {mean_p} AND population_c < {mean_p + std_derv}')
rows = c.fetchall()
count_pop = [row[0] for row in rows]
print (f'6 - How many countries are within 1 standard deviation of the mean: {len(count_pop)}')

conn.close()
```
