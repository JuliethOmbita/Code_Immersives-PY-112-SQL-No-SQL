### Class roster with mean test scores<a id="2"></a>
#### 30 points
#### Create a file called 'PYTHON_PTI_CAPSTONE_CLASS_ROSTER_GPA.TXT'
#### In should contain an unduplicate list of student names their graded scores and their mean
#### 1. Lastname, Firstname - Test Score 1, Test Score 2, ... Final Exam Score, Mean
#### 2. Lastname, Firstname - Test Score 1, Test Score 2, ... Final Exam Score, Mean
#### NOTE: The names should be in alphabetical order by last name, first name and in proper case
#### Use the file 'CAPSTONE_QUIZ_KEY_LESSON_1' through 5 to score the exams and calculate the mean

```python
# List of test keys => test_key
# List of students and greades => students

import glob
import os

proj_dir = r'/Users/julis_478/Documents/Python Immersives/PY-112 SQL, No SQL/Classes/Project/PYTHON_PTI_CAPSTONE_PROJECT_FILES'
os.chdir(proj_dir)

test_keys = []
students = []
for i in glob.glob('*.TXT'): # or for i in glob.glob('*.*'):
    igr,name, first_name,*ignore,lesson = i.replace('.TXT', '').split('_') #Get the test type and its keys 
    if name == 'QUIZ' or name == 'FINAL':     
        line = open(i, 'r') 
        line = line.readlines()
        question_score = int((100/len(line)))
        test_key = [name, lesson, question_score, line] # Test type, grades, question score
        test_keys.append(test_key) #List of test and final keys 
    else:
        std_name = f'{name.title()} {first_name.title()}'
        line = open(i, 'r') 
        line = line.readlines()
        student = [std_name, lesson, line] # Test type, grades, question score
        students.append(student) #List of test and final keys 
students = sorted(students)
test_keys = sorted(test_keys)


```

```python
#Test score 

scores = []
std = students[0][0] # First student name
grades = []
# Iterate over all the students and the key list. 
for x in range (len(students)):     #Iterate over students 
    for y in range (len(test_keys)):    #Iterate over test answers 
        # If its the same student name and lesson number then get the grade
        if std == students[x][0] and (students[x][1] == test_keys[y][1] or (students[x][1] == 'EXAM' and test_keys[y][1] == 'KEY')): 
                score = 0 
                for i in range(len (test_keys[y][-1])): #Iterate over each question 
                    if students[x][-1][i] == test_keys[y][-1][i]: # Iteration over question 
                        score += test_keys[y][2] #Quetion value
                scores.append(score)
        # Else reassing a new name and storage the information.
        elif not std == students[x][0] or (x == (len(students)-1) and y == (len(test_keys)-1)):
            mean = round(sum(scores)/len(scores),2)
            score_copy = scores.copy()
            grade = [std, score_copy, mean] 
            grades.append(grade) 
            std = students[x][0]
            scores = []
            # Printing student intomation
            print (std)
            for i, cont in enumerate (score_copy):
                if i == (len(score_copy) - 1):
                    print (f'\tFinal Exam: {cont}')
                else: 
                    print (f'\tExam {i+1}: {cont}')
            print(f'\tMean: {mean}')

# Whiting into file
proj_dir = r'/Users/julis_478/Documents/Python Immersives/PY-112 SQL, No SQL/Classes/Project'
os.chdir(proj_dir)
with open('PYTHON_PTI_CAPSTONE_CLASS_ROSTER_GPA.TXT', 'w') as f:
    headline = 'Student Name, Grades, Mean'
    f.write("%s\n" % headline)
    for grad in grades:
        f.write("%s\n" % grad)
    print ('\nWriting in the list of class, grades and mean into a .txt file...')
        
```

### Class summary report<a id="4"></a>
#### 15 points
#### Create a file called 'PYTHON_PTI_CAPSTONE_SUMMARY REPORT.TXT'
#### It should contain the following
#### 1. There were X students
#### 2. Exam 1 - mean
#### 3. Exam 2 - mean
#### 4. Exam 3 - mean
#### 5. Exam 4 - mean
#### 6. Exam 5 - mean
#### 7. Final exam - mean

```python
def calc_mean(myList):
    '''
    Calculate the mean of a list or a tuple
    return a float
    '''
    return round(sum(myList)/len(myList),2)
    
proj_dir = r'/Users/julis_478/Documents/Python Immersives/PY-112 SQL, No SQL/Classes/Project'
os.chdir(proj_dir)
with open('PYTHON_PTI_CAPSTONE_SUMMARY REPORT.TXT', 'w') as f:
    #Get the list of grades
    lst = [item[1] for item in grades] 
    msg_write = f'1. There are {len(lst)} students.'
    print (msg_write)
    f.write("%s\n" % msg_write)
    # Iteration over list of grades
    for i in range(len(lst[0])): 
        lst1 = [item[i] for item in lst]
        if i == len(lst[0]) - 1:
            msg_write = f'{i+2}. Final Exam {i+1}: {calc_mean(lst1)}'
            print (msg_write)
            f.write("%s\n" % msg_write)
        else:
            msg_write = f'{i+2}. Exam {i+1}: {calc_mean(lst1)}'
            print (msg_write)
            f.write("%s\n" % msg_write)
    print ('\nWriting in the report and mean into a .txt file...')
```
