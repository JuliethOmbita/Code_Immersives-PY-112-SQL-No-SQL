Mersenne Primes<br>
Named for the French Minim friar who studied them <br>
Are prime numbers that are 1 less than a power of 2<br>
$$
2^n-1 
$$
Find the Mersenne primes up to a power of 35

```python
num_primes = []
mersenne_primes = []
def mersenne_prime (limit_power):
    '''
    This function finds the Mersenne primes from 0 to the limiting power.
    '''
    for num in range (0,limit_power):
        flag = False
        # prime numbers are greater than 1
        if num > 1:
            # check for factors
            for i in range(2, num):
                if (num % i) == 0:
                    # if factor is found, set flag to True and break out of loop
                    flag = True
                    break
        # check if flag
            if not flag:
                num_primes.append(num)
                mersenne_primes.append((2**num) - 1)
            
limit_power = 35    
mersenne_prime (limit_power)

for i in range (len(num_primes)):
    print (f"The Mersenne prime number of {num_primes[i]} is {mersenne_primes[i]}.")
```

A twin prime is a pair of prime numbers separated by 2<br>
For example 3,5 and 5,7 and 11,13 are twin primes.<br>
Find all twin primes between 1000 and 3000

```python
num_primes = []
def is_prime (lower_limit, higher_limit):
    '''
    This function finds the prime numbers from the lower limit to the upper limit.
    '''
    for num in range (lower_limit, higher_limit):
        flag = False
        if num > 1:
            for i in range(2, num):
                if (num % i) == 0:
                    flag = True
                    break
            if not flag:
                num_primes.append(num)
            
def twin_primes(lower_limit, higher_limit):
    '''
    This function finds the twin primes from a list of numbers.
    '''
    is_prime(lower_limit, higher_limit)
    for i in range (len(num_primes)-1):
        if (num_primes[i+1]-num_primes[i]) == 2:
            print (f'The twin prime of {num_primes[i]}: {num_primes[i+1]}')

lower_limit = 1000
higher_limit = 3000  
twin_primes(lower_limit, higher_limit)
```

Fermat primes are prime numbers such that for the number n:<br>
$$
2^{n} + 1
$$
is also a prime number.  There are only 5 known fermat primes greater than 2.  They are all under 100,000

```python
def fermat_primes (num):
    '''
    This funtion finds the first 5 Fermat primes below 100.000
    '''
    print ('The fermant number if Fn= 2^2^n.\n')
    print ('The first 5 Fermat primes below 100.000 are:')
    for i in range (num):
        fermant_number = (2**(2**i) + 1) 
        print(f'\tF{i} = {fermant_number}')
num = 5
fermat_primes (num)
```
