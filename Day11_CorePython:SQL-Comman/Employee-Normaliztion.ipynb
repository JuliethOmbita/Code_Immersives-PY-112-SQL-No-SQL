```python
import sqlite3
import os
cwd = os.getcwd()
os.chdir(r'/Users/julis_478/Documents/Python Immersives/PY-112 SQL, No SQL/Classes/Day6')
cwd = os.getcwd()
conn = sqlite3.connect('ABC_corp.db')  # <== create the database
c = conn.cursor()

with open('ABC_CORP_DATA.csv','r') as myFile:

    # 1 - Add the data to a data structure
    headLine = myFile.readline()
    content = myFile.readlines()
    # print(content[0])
    # print(type(content[0]))
    for presd in (content):
        # Clean the data
        line_elements = presd.split(',')
        # print(line_elements[0],line_elements[1],line_elements[2])
        # print(line_elements[:13])
        c.execute('''INSERT   INTO employees
                VALUES (?,?,?,?,?,?,?,?,?,?,?,?,?)''',line_elements[:13])
        conn.commit()
    print('='*35)
    # Retreive all table names in the database
    for row in c.execute("""SELECT name FROM sqlite_master"""):
        print(row)
    print('='*35)
    for row in conn.execute("""SELECT  * FROM employees"""):
        print(row)
```
