# 1. Exercise - Even number. 
Find all numbers from 1000 to 3000 that are even numbers and when you add the digits the sum is divisible by 5 <br>
Also output the total number of items found for example 1004 is even and the sum of the digits is 5 which is divisible by 5

```python
def even(num):
    if num % 2 == 0:
        return True 
    else:
        return False 


for i in range (1000, 3000):
    if even(i):
        add = 0 
        for y in range(len(str(i))):
            add = add + int(str(i)[y])
        if add % 5 == 0 :
            print (f'This ia a even number {i}, its add is {add} and divisible by 5.')
```

```python
# Aternative Solution

for i in range(1000, 3001):
    if i % 2 == 0:
        sum_str = sum([int(x) for x in str(i)])
        if sum_str % 5 == 0:
            print(i, str(i), sum_str)
```
